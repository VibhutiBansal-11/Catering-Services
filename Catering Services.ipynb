```python
import pandas as pd
from datetime import date
#import display



drinks = pd.read_csv(r"Book2.csv")
foods = pd.read_csv(r"Book1.csv")
services = pd.read_csv(r"services.csv")
order =  pd.DataFrame(columns = ['item','cost'])
user = pd.DataFrame(columns =['name','event','guest','venue','date']) 

def food_order():
    global foods,order
    print(foods)
    conti = 'y'
    while ((conti == 'y' ) or (conti == 'Y')):
        choice = int(input("Enter your choice"))
        order = order.append({'item':foods.iloc[[choice],0].values,'cost':foods.iloc[[choice],1].values},ignore_index=True)
        conti = input("Do you want to enter more choices(y/n)")
        
def drinks_order():
    global drinks,order
    print(drinks)
    conti = 'y'
    while ((conti == 'y' ) or (conti == 'Y')):
        choice = int(input("Enter your choice"))
        order = order.append({'item':str(drinks.iloc[[choice],0].values),'cost':drinks.iloc[[choice],1].values},ignore_index=True)
        conti = input("Do you want to enter more choices(y/n)")
        
def services_order():
    global servies,order
    print(services)
    conti = 'y'
    while ((conti == 'y' ) or (conti == 'Y')):
        choice = int(input("Enter your choice"))
        order = order.append({'item':services.iloc[[choice],0].values,'cost':services.iloc[[choice],1].values},ignore_index=True)
        conti = input("Do you want to enter more choices(y/n)")

def bill():
    total = sum(order.iloc[:,1].values)
    discount = 0.05 #5 percent
    bill_amount = total - discount * total 
    print("The bill amount is ",total,"The discount is ",total*discount,"The bill amount is",bill_amount)
    
def payment():
    global user,order
    today = date.today()
    d4 = today.strftime("%b-%d-%Y")
    print("ORDER DATE", d4)
    print("The user details are")
    print(user)
    print(order)
    bill()
    
```

```python
def user_input():
    global user
    print(user)
    name = input("Enter your name") 
    event = input("What is the event") 
    guest = input("Enter number of guests") 
    venue = input("Enter venue details") 
    date = input("Enter date of event")
    user = user.append({'name': name,'event' : event , 'guest' : guest , 'venue' : venue,'date':date}, ignore_index=True)
```

```python
def def_order_menu(): 

    global order
    
    while True:     
        
        # While looping to keep menu alive
        print("*" * 31 + "ORDER PAGE" + "*" * 31 + "\n"     # Mail Menu
              "\t(F) FOOD\n"
              "\t(D) DRINKS\n"
              "\t(O) OTHER SERVICES\n"
              "\t(M) MAIN MENU\n"
              "\t(E) EXIT\n" +
              "_" * 72)

        input_1 = str(input("Please Select Your Operation: ")).upper() # Options Handling : F-O-M-E.
        if len(input_1) == 1:
            if (input_1 == 'F'):  #Easy Access Checking Logic
                print("\n" * 10)
                food_order()
            elif (input_1 == 'D'):
                print("\n" * 10)
                drinks_order()
                
            elif (input_1 == 'O'):
                print("\n" * 10)
                services_order() # Show Services Menu
                
            elif (input_1 == 'M'):
                print("\n" * 10)
                main() # Show Main Menu
                
            elif (input_1 == 'E'):
                print("*" * 32 + "THANK YOU" + "*" * 31 + "\n")
                break
            
            else:
                print("\n" * 10 + "ERROR: Invalid Input (" + str(input_1) + "). Try again!") # Handling Bad Inputs
           
        else:
            print("\n" * 10 + "ERROR: Invalid Input (" + str(input_1) + "). Try again!")
    
```

```python
def main():
    global order
    while True:                                            #Repeat Menu until stops.
        print("*" * 31 + "MAIN MENU" + "*" * 32 + "\n"     #Design for Main Menu.
              "\t(U) USER INPUT\n"
              "\t(O) ORDER\n"
              "\t(R) REPORT\n"
              "\t(P) PAYMENT\n"
              "\t(E) EXIT\n" +
              "_" * 72)

        input_1 = str(input("Please Select Your Operation: ")).upper()    #Input, have to choose operation. Make everything UPPER symbol.
        if (len(input_1) == 1):                                           #Checking input length.
            if (input_1 == 'O'):                                          #If input is "O".
                print("\n" * 10)                                         #Create 100 empty lines.
                def_order_menu()
               
            elif (input_1 == 'R'):                                        #If input is "R".
                print("\n" * 10)                                         #Create 100 empty lines.
                
                bill()
                #Stop repeating Main Menu.
            elif (input_1 == 'U'):                                        #If input is "R".
                print("\n" * 10) 
                user_input()                                        #Create 100 empty lines.
            elif (input_1 == 'P'):                                        #If input is "P".
                print("\n" * 10)                                         #Create 100 empty lines.
                payment()                                           #Start Payment function.
                                                                     #Stop repeating Main Menu.
            elif (input_1 == 'E'):                                        #If input is "E".
                print("*" * 32 + "THANK YOU" + "*" * 31 + "\n")           #Good bye comment.
                break                                                     #Stop repeating Main Menu.
            else:                                                                                 #If O, R, P, E not inserted then...
                print("\n" * 10 + "ERROR: Invalid Input (" + str(input_1) + "). Try again!")     #Invalid input.
        else:                                                                                     #If input length not equal to 1...
            print("\n" * 10 + "ERROR: Invalid Input (" + str(input_1) + "). Try again!")         #Invalid input.'''
            
main()
```
